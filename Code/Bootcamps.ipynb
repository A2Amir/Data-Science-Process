```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('./survey_results_public.csv')
df.head()
print(df.shape)
```

#### To narrow our dataset down to just those individuals who are in bootcamp and not bootcamp. 

```python
bootcamp_df=df[df['TimeAfterBootcamp'].isnull()==False]
not_bootcamp_df=df[df['TimeAfterBootcamp'].isnull()==True]
```

```python
bootcamp_df.shape
```

### Looking at gender of those who are in bootcamp

```python
bootcamp_df['Gender'].value_counts()/(bootcamp_df.shape[0]-sum(bootcamp_df['Gender'].isnull()))
```

### Looking at gender of those who are not in bootcamp

```python
not_bootcamp_df['Gender'].value_counts()/(not_bootcamp_df.shape[0]-sum(not_bootcamp_df['Gender'].isnull()))
```

By comparing: 

* There is about a four percent larger group in the males that are not in boot camps
* There is a  increase of females in boot camps

### look at a breakdown by formal education. 

```python
bootcamp_df['FormalEducation'].value_counts()/(bootcamp_df.shape[0]-sum(bootcamp_df['FormalEducation'].isnull()))
```

* Of those who were in the boot camp,about 48 percent had a bachelor's degree before going in,and 23 percent have a master's degree. 

### Formal education of the people who are not in a boot camp, 

```python
not_bootcamp_df['FormalEducation'].value_counts()/(not_bootcamp_df.shape[0]-sum(not_bootcamp_df['FormalEducation'].isnull()))
```

* You can see that fewer of them have a bachelor's degree,about the same have a master's degree, 

### Another interesting column that we could look at is basically how much time after the boot camp it takes for an individual to find a job. 

```python
bootcamp_df['TimeAfterBootcamp'].value_counts()/(bootcamp_df.shape[0]-sum(bootcamp_df['TimeAfterBootcamp'].isnull()))
```

* 47 percent of those individuals already had jobs
* 9 percent got them immediately after graduating 
* about 10 percent got jobs as developers before completing the program.
